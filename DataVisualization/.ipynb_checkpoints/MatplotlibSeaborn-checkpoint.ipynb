```python
# Hİpotez Testi = Özellikle üniversitelerde anketler ile o hipotezin doğruluğunu kontrol etmek için kullanılır.
# H0 = null hipotezi, yokluk hipotezi (İki grup arasında hiçbir fark yoktur anlamına gelmektediir.)
# Alternative Hipotez = İki grup arasında fark vardır anlamına gelmektedir.(Online ders alanlar daha başarılıdır, klasik ders alanlar 
# başarısızdır gibi olabilir.)
# Örnek = C vitamini alan annelerin doğum sırasında bir fark vardır veya yoktur. H0= C vitamini alan veya almayan arasında fark yoktur. 
# Aletnative Hipotez= C vitamini alanlar daha çok kız doğurmuştur.
# Hipotez Testi İş Akışı:
# Araştırma Problemi: "Yeni ilaç eski tedaviden daha etkili midir?"
# Hipotezler:
# H₀: Yeni ilacın etkisi, eski tedavi ile aynıdır.
# H₁: Yeni ilaç, eski tedaviden daha etkilidir.
# Test Seçimi: İki grup ortalaması karşılaştırılacaksa bağımsız örneklem t-testi seçilebilir.
# Varsayımların Kontrolü: Verilerin normal dağıldığından emin olunur.
# Anlamlılık Düzeyi (α): %5 olarak belirlenir.
# Testin Uygulanması: Test istatistiği ve p-değeri hesaplanır.
# Karar Verme: p ≤ 0.05 ise H₀ reddedilir; yeni ilacın etkili olduğuna dair kanıt vardır.
# Sonuçların Yorumlanması: Test sonuçları raporlanır ve klinik anlamlılığı da dikkate alınır.
```

```python
# İndependent(input/giriş) and dependent(output/çıkış) değişkenlerler
# Bağımsız değişken,diğer verilerden etkilenmesi gbeklemnen vveriyi tanımlar. C vitamini takviyesi, Hava durumu kapalı yağmur var kar var
# Bağımlı değişken, bağımsız değişken veya değişkenler tarafından etkilenen değişlendir. Doğum cinsiyet oranı, maç oynanacak oynanmayacak(hava durumuna bağlı)
# Pyhtonda X(input) y(output) olarak gösterilir.
```

```python
# Desing of Experimant
# Veriler genellikle belirli bir soruyu yanıtlamayı amaçlayan bir çalışmanının sonucu olarak oluşturulur.ANcak verilerin nasıl 
# oluşturulduğuna ve çalışmanının
# nasıl oluştuğuna ve çalışmanının nasıl tasarandığına bağlı olarak verilen farklı şekilde analiz edilmesi ve yorumlanması gerekir.
# Controlled experiments (Kontrollü denemeler)
# Bir gruba normal ders bir gruba sanal gerçeklik dersi anlatılıyor. h0= iki grup da aynı başarıyı elde eder. 
# H1= sanal gerçeklik alanlar daha başarılıdır.
# Burada popülasyon 2024 2025 yılında ilk defa bu dersi alan öğrencilerdir.
# Deneylerin altın standardı
# KOntrol grubu eskiden devam ederken uygulam grubu yeni bir yol dener. Bu sayede yeni yol iyi mi kontrol edilir. A B testidir
# Correlation: 
# İki değilken arasındaki ilişkiyi ölçmenin yoludur.
# Bu katsayı -1 ile 1 arasındadır. (1 doğru orantı, -1 ters orantı)
# Eğer correnation 0 ise ikisi arasında ilişki yoktur
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df_sleep=pd.read_csv('msleep.csv')
sns.scatterplot(x='sleep_total',y='sleep_rem',data=df_sleep) # Iki sutun arasindaki iliskiyi gosteren dagilim grafigini olusturur.
```

```python
df_sleep['sleep_total'].corr(df_sleep['sleep_rem']) # Iki sutun arasindaki kolerasyon katsayisini hesaplar.
# Sonuc 1`e yakin pozitif bir deger ciktigindan aralarinda guclu bir iliski vardir diyebiliriz. (Birbirlerine bagimlilar.)
```

```python
sns.lmplot(x='sleep_total',y='sleep_rem',data=df_sleep,ci=None) # Dogrusal regresyon cizgisiyle iki sutun arasindaki iliskiyi gosterir.
# ci=None sadece veri noktasini ve regresyon cizgisini gosterir.
plt.show()
```

```python
df_sleep.corr(numeric_only=True) # numeric_only=True sadece sayisal sutunlari secer.
# Bu kod sutunlarin birbirleri arasindaki iliskiyi gosteren korelasyon matrisini olusturur.
```

```python
sns.heatmap(df_sleep.corr(numeric_only=True)) # Ustteki kodun isi haritasi.
```

```python
sns.heatmap(df_sleep.corr(numeric_only=True),annot=True) # Ustteki grafigin her bir hucresine korelasyon katsayisinin yazilmis hali.
```

```python
sns.heatmap(df_sleep.corr(numeric_only=True),annot=True,cmap='coolwarm') # Renk paletinin degistirilmis hali.
```

```python
sns.heatmap(df_sleep.corr(numeric_only=True),annot=True,cmap='crest')
```

```python
import pandas as pd
df_coffee=pd.read_feather('coffee_ratings_full.feather')
df_coffee
```

```python
# sistematik örnekleme
# veri setindeki satır sayısı = 1338 satır
# sayısal örneklem yüyüklüğü = 5
# aralık 1338//5 = 267
```

```python
samp_size=5
pop_size=df_coffee.shape[0] # Bu kod df in kac tane satiri oldugunu bulmamizi saglar. shape kodu normalde hem satiri hem sutunu 
# gosterdiginden oturu indexi 0 girerek sadece satiri gostermesini saglar.
pop_size2=len(df_coffee) # df in kac satiri oldugunu gosterir.
pop_size
pop_size2
```

```python
interval=pop_size//samp_size # Toplam populasyon sayisini ornek populasyin sayisina bolerek ornekleme yaparken kullanilacak bir aralik 
# olusturmamizi saglar.
interval
```

```python
# Systematik sampling -selecting the rows (Sistematik Ornekleme ile Satir Secimi)
df_coffee.iloc[interval::interval] # Buldugumuz interval satirina denk gelen satiri baslangic olarak kabul ederek interval sayisi kadar 
# atlayip denk gelen satirlari df olarak ekrana yazdirir.
```

```python
# Yazilan Kodlarin Toplu Hali:
kahveUlkeSayisi=df_coffee['country_of_origin'].value_counts() # Hangi ulkenin kac farkli cesit kahvesi var.
ilkAltiUlke=kahveUlkeSayisi.head(6).index # En cok kahve cesidi olan 6 ulke.
# Burada index i uyusup uyusmadigina direkt ulke adindan bakmak icin kullaniyoruz.
# ilkAltiUlke=list(df_coffee.groupby('country_of_origin')['country_of_origin'].agg('count').nlargest(6).index) # Ikinci Yol
uyusanUlkeler=df_coffee['country_of_origin'].isin(ilkAltiUlke) # Asil dfte ilk 6 ulkenin verilerinin yer alip almadigini karsilastirir.
ilkAltiUlkeUyusanlar=df_coffee[uyusanUlkeler] # Uyusanlardan yeni bir df olusturur.
ilkAltiUlkeUyusanlarOrnek=ilkAltiUlkeUyusanlar.sample(frac=0.1,random_state=2024) # df in yuzde 10 luk kismindan rastgele veriler secerek
# yeni bir df olusturur.
# ilkAltiUlkeUyusanlarOrnek=ilkAltiUlkeUyusanlar.groupby('country_of_origin').sample(frac=0.1,random_state=2024) # Ikinci Yol
ilkAltiUlkeUyusanlarOrnek2=ilkAltiUlkeUyusanlar.groupby('country_of_origin').sample(n=15,random_state=2024) # Her ulkeden esit adette
# rastgele deger secilmesini saglar.
ilkAltiUlkeUyusanlarOrnek['country_of_origin'].value_counts(normalize=True) # Ornek df te yer alan ulke adlarinin oranini yuzdelik olarak
# yazdirir.
ilkAltiUlkeUyusanlarOrnek2['country_of_origin'].value_counts() # Ornek dfte yer alan ulkeleri kendileri arasinda toplar.
```

```python
coffee_rating_weight=coffee_ratings_top
condition=coffee_rating_weight['country_of_origin']=='Taiwan' # Bu kodla coffee_rating_weight df inin country_of_origin sutunundaki degerler
# Taiwan ise True degilse False donduruyoruz.

import numpy as np
coffee_rating_weight['weight']=np.where(condition,2,1) # Bu kodla weight adinda yeni bir sutun olusturarak hazirdaki df e ekleriz. weight
# sutunundaki degerler condition sartina uyuyorsa True yani 2 uymuyorsa False yani 1 degerini alir.
coffee_rating_weight=coffee_rating_weight.sample(frac=0.1,weights='weight') # Bu orneklem olusturmada weight sutunundaki yuksek degerli 
# satirlar oncelikli secilir.
coffee_rating_weight['country_of_origin'].value_counts(normalize=True)
```

```python
# cluster sampling
# örneği toplamak için fiziksel bir yere seyehat edilmesi gerektiğinde aşırı pahalı hale gelir bunun yerine ucuz hale getirmek için cluster 
# samplig kullanılır
```

```python
varieties_pop=list(df_coffee['variety'].unique()) # variety sutunundaki benzersiz degerleri secip liste olusturarak bu listeye varieties_pop
# adini verir.
varieties_pop
```

```python
import random
varieties_samp=random.sample(varieties_pop,k=3) # varieties_pop list inden 3 farkli degeri secerek varieties_samp e aktarir.
varieties_samp
```

```python
pd.set_option('display.max_columns',None)
veriety_condition=df_coffee['variety'].isin(varieties_samp)
coffee_ratings_cluster=df_coffee[veriety_condition]
coffee_ratings_cluster
```

```python
coffee_ratings_cluster.loc[:,'variety']=coffee_ratings_cluster['variety'].astype('category') # loc ile variety sutununun butun satirlarina eriserek
# bu sutuna ait olan satirlardaki ifadelerin category veri tipine donusturulmesini saglar.
# loc[baslangicSatiri(Dahil),BitisSatiri(Dahil),hangiSutununSatiri]  
# astype ile veri tipi degisimi yapiyoruz.
coffee_ratings_cluster['variety']=coffee_ratings_cluster['variety'].cat.remove_unused_categories() # Verilen sutunda mevcut olan kategorileri
# tutup kalanini siler.
```

```python
coffee_ratings_cluster.groupby('variety',observed=True).sample(n=5,random_state=2024,replace=True) # Bu kodun amaci variety sutununu baz
# alip observed=True ile mevcut olan kategorilerden her farkli degerden 5 tane olacak ve kod her calistiginda bu degerler degismeyecek 
# sekilde cikti vermektir.
# Buradaki sikinti su her veriden 5 farkli deger ver dedigimizde 5 farkli degeri olmayan vari oldugundan kod hata veriyor. Bu yuzden replace=
# True ile ayni degerlerin secilmesini sagladim. Ama hocanin kodunda replace yer almiyor.
```

```python
# Multistage sampling(çok aşamalı örnekleme)
# ikiden fazla aşama kullanmak müümkündür. örneğin eyalet ilçe şehir mahalle
```

```python
coffee_ratings_srs=coffee_ratings_top.sample(frac=1/3,random_state=2024)
coffee_ratings_srs.shape
# Veri setinden yuzde 33 luk bir kismi rastgele alarak yeni bir df olusturup sonrasinda bu df in satir ve sutun sayisini yazdirdik.
```

```python
coffee_ratings_strat=coffee_ratings_top.groupby('country_of_origin').sample(frac=1/3,random_state=2024)
coffee_ratings_strat.shape
# frac kismina girdigimiz deger df in belli bir kisminda deger almamiz gerektigini belirtmez. Sadece df in tamamindan girdigimiz kadar deger 
# almamizi saglar.
# Bu kod ve ustteki kodun sonuclari ayni da cikabilir ama farkli cikma olasiligi cok yuksek. Hocada farkli cikarken bende ayni cikti.
# Bu iki kodun arasindaki mantik ustte tum df uzerinda islem yaptigimizdan yani elimizdeki veri fazla oldugundan secilen satir ve sutun
# sayisi, bu kodda direkt ulke adina secim yaptigimizdan oturu cok daha fazladir, bunu karsilastiriyoruz.
```

```python
import random
top_countries_sample=random.sample(top_counted_countries,k=2)
top_condition=coffee_ratings_top['country_of_origin'].isin(top_countries_sample)
coffee_ratings_cls=coffee_ratings_top[top_condition]
# coffee_ratings_cls.loc[:,'country_of_origin']=coffee_ratings_cls.astype('category').cat.remove_unused_categories()
coffee_ratings_clst=coffee_ratings_cls.groupby('country_of_origin').sample(len(coffee_ratings_top)//6,replace=True)
coffee_ratings_clst.shape
```

```python
# P değeri için p<= 0.05 H0 red edilir
# p>0.05 H0 bakul edilir
# iki normal dağılımda iki dağılım arasında kalan alana significance level denir.
# Null hiponetiz true iken reddediyorsak tip 1 error almış oluruz
# Null hipotezi false iken kabul ediyorsak tip 2 error almış oluruz
```

```python
# Beta dağılımı iki parametre ile tanımlalan sürekli bir dağılımdır 
# Eğer a=b=1 ise dağılım uniform(eşit olasılıklı) dağılım olur
# Eğer a>1 ve b>1, dağılım daha merkezi hale gelir ve 0 ile 1 arasında yoğunlaşır.
# Eğer a<1 ve b<1ise, dağılım uç noktalarda yoğunlaşır(yani 0 ve 1'e yakın olur)
```

```python
randoms=np.random.beta(a=2,b=2,size=5000) # a ve b dagilimlari 2 olan yani dagilimi ortada (0.5) olan beta dagilimindan rastgele 5000 adet
# veri olusturur.
randoms
```

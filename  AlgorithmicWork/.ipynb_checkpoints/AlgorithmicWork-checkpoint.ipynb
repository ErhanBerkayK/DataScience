```python
import pandas as pd
df_animal=pd.read_csv('msleep.csv')
df_animal
```

```python
### 1. df Uzerindeki Tum Aykiri Degerleri Bulma : 
```

```python
import numpy as np
from scipy.stats import iqr
iqr_deger=iqr(df_animal.sleep_total)
alt_sinir=np.percentile(df_animal.sleep_total,0.25)-1.5*iqr_deger
ust_sinir=np.percentile(df_animal.sleep_total,0.75)+1.5*iqr_deger
sonuc_df=df_animal[(df_animal.bodywt<alt_sinir) | (df_animal.bodywt>ust_sinir)]
sonuc_df
```

```python
### 2. Bir Seri Olusturup O Seriden Rastgele Degerleri Baska Bir Seriye Atayip O Verilerin Ortalamasini Alip Grafigini Cizdirme :
```

```python
import matplotlib.pyplot as plt
liste=pd.Series([1,2,3,4,5,6])
bos_liste=[]
for i in range(100000):
    ornek_liste=liste.sample(n=10,replace=True)
    bos_liste.append(ornek_liste.mean())
plt.title('')
plt.hist(bos_liste)
```

```python
### 3. Bir Seri Olusturup O Seriden Rastgele Degerler Uretip Rastgele Degerlerden Istedigimiz Degerle Eslesen Degerin Orneklem Uzerindeki
### Oranini Hesaplayip Grafigini Cizdirme :
```

```python
liste=pd.Series(['Erhan','Berkay','Kirazoglu'])
bos_liste=[]
for i in range(100000):
    ornek_liste=liste.sample(n=10,replace=True)
    try:
        bos_liste.append(ornek_liste.value_counts()['Berkay']/10)
    except:
        bos_liste.append(0)
plt.title('')
plt.hist(bos_liste)
```

```python
### 4. df teki Bir Sutunun Logaritmasini Alip df e Yeni Bir Sutun Olarak Ekleyip Baska Bir Sutunla Korelayon Hesabini Yaptirip Grafigini 
### Cizme : 
```

```python
import seaborn as sns
df_animal['log_bodywt']=np.log(df_animal.bodywt)
print(df_animal.log_bodywt.corr(df_animal.awake))
sns.lmplot(x='log_bodywt',y='awake',data=df_animal)
```

```python
### 7. df teki En Buyuk 6 Degeri Eski df le Karsilastirip Uyusanlari Yeni bir df e Atip Bu df Uzerinde Rastgele Orneklemler Elde Etme ve Bu 
### Degerler Uzerinde Karsilastirmalar Yapma : 
```

```python
import random
ilkAltiUlke=list(df_coffee.groupby('country_of_origin')['country_of_origin'].count().sort_values(ascending=False).head(6).index)
ilkAltiUlkeKarsilastirma=df_coffee['country_of_origin'].isin(ilkAltiUlke)
ilkAltiUlkeUyusanlar=df_coffee[ilkAltiUlkeKarsilastirma]
ilkAltiUlkeUyusanlarOrnek=random.sample(ilkAltiUlke,k=2)
ilkAltiUlkeKarsilastirma2=ilkAltiUlkeUyusanlar['country_of_origin'].isin(ilkAltiUlkeUyusanlarOrnek)
ilkAltiUlkeUyusanlar2=ilkAltiUlkeUyusanlar[ilkAltiUlkeKarsilastirma2]
ilkAltiUlkeUyusanlar2.loc[:,'country_of_origin']=ilkAltiUlkeUyusanlar2['country_of_origin'].astype('category').cat.remove_unused_categories()
ilkAltiUlkeUyusanlar2Ornek=ilkAltiUlkeUyusanlar2.sample(len(ilkAltiUlkeUyusanlar)//6)
ilkAltiUlkeUyusanlar2Ornek2=ilkAltiUlkeUyusanlar2.groupby('country_of_origin').sample(len(ilkAltiUlkeUyusanlar)//6,replace=True)
ilkAltiUlkeUyusanlar2Ornek.shape
ilkAltiUlkeUyusanlar2Ornek2.shape
```

```python
### 9. Yeni Bir df Olusturup Onceki df Uzerinde Orneklemler Olusturup Goreceli Hata Hesaplayip Yeni df e Aktarma :
```

```python
df_hata=pd.DataFrame(columns=['orneklem_buyuklugu','hata_degeri'])
populasyon_ortalamasi=df_coffee['total_cup_points'].mean()
for satir_sayisi in range(1,len(df_coffee)+1):
    orneklem_ortalamasi=df_coffee.sample(n=satir_sayisi)['total_cup_points'].mean()
    hata=np.abs(populasyon_ortalamasi-orneklem_ortalamasi)/populasyon_ortalamasi
    df_hata.loc[len(df_hata.index)]=[satir_sayisi,hata]
df_hata.plot(x='orneklem_buyuklugu',y='hata_degeri',kind='line')
df_hata
```
